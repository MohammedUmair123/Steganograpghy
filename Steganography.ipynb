```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import cv2
import os

class SteganographyApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Secure Data Hiding in Images Using Steganography")
        self.root.geometry("700x600")

        self.img = None
        self.password = ""
        self.message = ""

        # UI elements
        self.label_image = tk.Label(self.root, text="No image selected")
        self.label_image.pack()

        self.select_image_button = tk.Button(self.root, text="Select Image", command=self.select_image)
        self.select_image_button.pack()

        self.image_label = tk.Label(self.root)
        self.image_label.pack()

        self.message_label = tk.Label(self.root, text="Enter Secret Message:")
        self.message_label.pack()

        self.message_entry = tk.Entry(self.root, width=40)
        self.message_entry.pack()

        self.password_label = tk.Label(self.root, text="Enter Passcode:")
        self.password_label.pack()

        self.password_entry = tk.Entry(self.root, width=40, show="*")
        self.password_entry.pack()

        self.encrypt_button = tk.Button(self.root, text="Encrypt Image", command=self.encrypt)
        self.encrypt_button.pack()

        self.decrypt_button = tk.Button(self.root, text="Decrypt Image", command=self.decrypt)
        self.decrypt_button.pack()

        self.decryption_label = tk.Label(self.root, text="Enter Passcode to Decrypt:")
        self.decryption_label.pack()

        self.decryption_entry = tk.Entry(self.root, width=40, show="*")
        self.decryption_entry.pack()

        self.encrypted_image_label = tk.Label(self.root, text="Encrypted Image will appear here")
        self.encrypted_image_label.pack()

    def select_image(self):
        file_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.jpg;*.jpeg;*.png")])
        if file_path:
            self.img = cv2.imread(file_path)
            self.display_image(self.img)  # Display the selected image

    def resize_image(self, image, max_width=300, max_height=300):
        # Resize the image to fit within max_width and max_height
        img_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        img_pil.thumbnail((max_width, max_height))  # Resize with aspect ratio preserved
        return img_pil

    def display_image(self, image):
        # Resize and display the image
        img_pil = self.resize_image(image)
        img_tk = ImageTk.PhotoImage(img_pil)  # Convert to Tkinter-compatible image
        self.image_label.config(image=img_tk)
        self.image_label.image = img_tk  # Keep a reference to the image object

    def encrypt(self):
        if self.img is None:
            messagebox.showerror("Error", "No image selected")
            return

        self.message = self.message_entry.get()
        self.password = self.password_entry.get()

        if not self.message or not self.password:
            messagebox.showerror("Error", "Message and passcode are required")
            return

        d = {}
        c = {}
        for i in range(255):
            d[chr(i)] = i
            c[i] = chr(i)

        m = 0
        n = 0
        z = 0

        for i in range(len(self.message)):
            self.img[n, m, z] = d[self.message[i]]
            n += 1
            m += 1
            z = (z + 1) % 3

        encrypted_image_path = "encryptedImage.jpg"
        cv2.imwrite(encrypted_image_path, self.img)
        os.system(f"start {encrypted_image_path}")

        # Display the encrypted image in GUI
        encrypted_img = cv2.imread(encrypted_image_path)
        self.display_encrypted_image(encrypted_img)

        messagebox.showinfo("Success", "Image encrypted successfully!")

    def display_encrypted_image(self, image):
        # Resize and display the encrypted image
        img_pil = self.resize_image(image)
        img_tk = ImageTk.PhotoImage(img_pil)  # Convert to Tkinter-compatible image
        self.encrypted_image_label.config(image=img_tk)
        self.encrypted_image_label.image = img_tk  # Keep a reference to the image object

    def decrypt(self):
        if self.img is None:
            messagebox.showerror("Error", "No image selected")
            return

        pas = self.decryption_entry.get()
        if pas != self.password:
            messagebox.showerror("Error", "Invalid passcode")
            return

        message = ""
        n = 0
        m = 0
        z = 0
        c = {}
        for i in range(255):
            c[i] = chr(i)

        for i in range(len(self.message)):
            message += c[self.img[n, m, z]]
            n += 1
            m += 1
            z = (z + 1) % 3

        messagebox.showinfo("Decryption", f"Decrypted Message: {message}")


if __name__ == "__main__":
    root = tk.Tk()
    app = SteganographyApp(root)
    root.mainloop()
```
